```python
import numpy as np
import pandas as pd
import tensorflow as tf
import keras
from keras import models,layers
```

```python
dataset = pd.read_csv("report_till_2021.csv")
```

```python
dataset.head()
```

```python
dataset.drop(columns={"Unnamed: 0","Country name","year"},axis=1,inplace=True)
```

```python
dataset.head()
```

```python
corr_matrix = dataset.corr()
```

```python
corr_matrix["Life Ladder"].sort_values(ascending=False)
```

```python
X = dataset.iloc[:,1:].values
y = dataset.iloc[:,0].values
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="mean")
imputer.fit(X)
X = imputer.transform(X)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from keras.layers import Dense,Dropout
ann = models.Sequential()
ann.add(Dense(units=6,activation="relu",input_dim=6))
ann.add(Dropout(0.2))
ann.add(Dense(units=6,activation="relu"))
ann.add(Dropout(0.2))
ann.add(Dense(units=3,activation="relu"))
ann.add(Dense(units=1,activation="sigmoid"))
```

```python
ann.compile(optimizer="adam",loss="mean_squared_error")
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
y_pred = ann.predict(X_test)
```

```python
error_rate = 0
for i in range(len(y_pred)):
    error = abs(y_pred[i] - y_test[i])
    error = error / y_test[i]
    error_rate += error
error_rate /= len(y_pred)
```

```python
accuracy = 100 - error_rate
accuracy
```
